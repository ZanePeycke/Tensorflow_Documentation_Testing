```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

```python
# define and compile the NN
# This network has one layer, and that layer has 1 neuron
# The input shape is 1 value
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

```python
# Now we compile, we specify loss function and an optimizer
# We make a prediction, use the loss function to measure how accurate this prediction is
# using previously seen data

# Then using the optimizer we make another guess by trying to minimze the loss
# We repeat this process for a number of epochs

model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
# Load in our data
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
yx = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

```python
#Training the neural network
# In training we learn the relationship between X and Y (during model.fit)
# make a guess, measure how bad the guess is, use the optimizer to make another guess

model.fit(xs, yx, epochs = 500)
```

```python
# Now we understand the relationship, we can predict for previously unseen data
print(model.predict([10.0]))
```
